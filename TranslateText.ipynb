```
from google.cloud import translate
import six
import sys

import os
os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = "/content/drive/My Drive/Colab Notebooks/TranslationService.json"

!export GOOGLE_APPLICATION_CREDENTIALS="/content/drive/My Drive/Colab Notebooks/TranslationService.json"
translate_client = translate.Client()

def myTranslate(text, source='de', target='en'):
  if isinstance(text, six.binary_type):
      text = text.decode('utf-8')

  # Text can also be a sequence of strings, in which case this method
  # will return a sequence of results for each text.
  result = translate_client.translate(
      text, target_language=target, source_language=source)
  return result['translatedText']

#print(u'Text: {}'.format(result['input']))
#print(u'Translation: {}'.format(result['translatedText']))
#print(u'Detected source language: {}'.format(
    #result['detectedSourceLanguage']))
```

```
f = open(file_path_in, "r")

from lxml import etree
tree = etree.parse(f)

for node in tree.xpath('.//mxCell'):
  if 'value' in node.attrib:
    node.attrib['value'] = myTranslate(node.attrib['value'])
    

xml = etree.tostring(tree, pretty_print=True)
f = open(file_path_out, "wb")
f.write(xml)
f.close()
```
