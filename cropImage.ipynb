```
import numpy as np


# 1024 pixels by 1024 pixels suggested maximum.

img = cv2.imread("/content/drive/My Drive/Colab Notebooks/raw/raw/links/20190828-093312.jpg")
crop_img = img[0:520, 0:960]
crop_img2 = img[0:520, 960:1920]

full_img = np.zeros(shape=[1040, 960, 3], dtype=np.uint8)
full_img[0:520, 0:960] = crop_img
full_img[520:1040, 0:960] = crop_img2

plt.imshow(full_img), plt.title("combined")
plt.show()
plt.imshow(img), plt.title("original")
plt.show()
```

```
in_path = "/content/drive/My Drive/Colab Notebooks/raw/raw/rechts/"
out_path = "/content/drive/My Drive/Colab Notebooks/raw/cropped/rechts/"

file_path = glob.glob(in_path)
ls_result = !ls "/content/drive/My Drive/Colab Notebooks/raw/raw/rechts/"
files = []
for items in ls_result:
  files += items.split()
 
print(files)

for file in files:
  img = cv2.imread(in_path + file)
  crop_img = img[0:520, 0:960]
  crop_img2 = img[0:520, 960:1920]

  full_img = np.zeros(shape=[1040, 960, 3], dtype=np.uint8)
  full_img[0:520, 0:960] = crop_img
  full_img[520:1040, 0:960] = crop_img2
  cv2.imwrite(out_path + file, full_img )
  print(in_path + file)
  
print("Finished")
  
```
